# Configurazione e Setup dell'Ambiente

Questo notebook configura l'ambiente di sviluppo installando e verificando i pacchetti dal file requirements.txt. 

Questo assicura che tutte le librerie necessarie per il progetto di deep learning siano installate correttamente e funzionino come previsto.

## 1. Installazione Dipendenze dal File Requirements

Prima di tutto, installiamo tutti i pacchetti specificati nel file requirements.txt, incluso il pacchetto locale in modalit√† development.

```python
import subprocess
import sys

# Installa le dipendenze dal file requirements.txt
def installa_dipendenze():
    """Installa i pacchetti dal file requirements.txt"""
    try:
        # Installa dal requirements.txt
        risultato = subprocess.run([
            sys.executable, "-m", "pip", "install", "-r", "../requirements.txt"
        ], capture_output=True, text=True, check=True)
        
        print("‚úÖ Tutti i pacchetti installati con successo!")
        print(risultato.stdout)
        
    except subprocess.CalledProcessError as e:
        print("‚ùå Errore nell'installazione dei pacchetti:")
        print(e.stderr)

# Esegui l'installazione
installa_dipendenze()
```

## 2. Verifica Installazione Pacchetti

Verifichiamo che tutti i pacchetti richiesti siano stati installati correttamente tentando di importarli.

```python
def verifica_installazione_pacchetti():
    """Verifica se tutti i pacchetti richiesti possono essere importati"""
    pacchetti = [
        'click', 'sphinx', 'flake8', 'dotenv', 
        'numpy', 'pandas', 'matplotlib', 'seaborn', 
        'sklearn', 'torch', 'torchvision', 'tensorflow',
        'tensorflow_datasets', 'jupyter'
    ]
    
    import_falliti = []
    
    for pacchetto in pacchetti:
        try:
            if pacchetto == 'sklearn':
                __import__('sklearn')
            elif pacchetto == 'dotenv':
                __import__('dotenv')
            elif pacchetto == 'tensorflow_datasets':
                __import__('tensorflow_datasets')
            else:
                __import__(pacchetto)
            print(f"‚úÖ {pacchetto}")
        except ImportError as e:
            print(f"‚ùå {pacchetto}: {e}")
            import_falliti.append(pacchetto)
    
    if import_falliti:
        print(f"\n‚ùå Import falliti: {import_falliti}")
    else:
        print(f"\n Tutti i {len(pacchetti)} pacchetti importati con successo!")

verifica_installazione_pacchetti()
```

## 4. Controllo Versioni delle Librerie

Visualizziamo le versioni di tutti i pacchetti installati per verificare che soddisfino i requisiti minimi specificati.

```python
import pkg_resources
from packaging import version

def controlla_versioni():
    """Controlla le versioni dei pacchetti installati rispetto ai requisiti"""
    
    # Definisce i requisiti minimi
    requisiti = {
        'numpy': '1.24.0',
        'pandas': '2.0.0', 
        'matplotlib': '3.7.0',
        'seaborn': '0.12.0',
        'scikit-learn': '1.3.0',
        'torch': '2.0.0',
        'torchvision': '0.15.0',
        'tensorflow': '2.13.0',
        'tensorflow-datasets': '4.9.0',
        'click': '8.1.7',
        'sphinx': '7.2.6',
        'flake8': '7.0.0'
    }
    
    print("Controllo Versioni Pacchetti:")
    print("=" * 55)
    
    for pacchetto, versione_min in requisiti.items():
        try:
            if pacchetto == 'scikit-learn':
                import sklearn
                versione_installata = sklearn.__version__
            elif pacchetto == 'tensorflow-datasets':
                import tensorflow_datasets as tfds
                versione_installata = tfds.__version__
            else:
                versione_installata = pkg_resources.get_distribution(pacchetto).version
            
            # Confronta le versioni
            if version.parse(versione_installata) >= version.parse(versione_min):
                stato = "‚úÖ"
            else:
                stato = "‚ö†Ô∏è"
                
            print(f"{stato} {pacchetto:<20} {versione_installata:<10} (min: {versione_min})")
            
        except pkg_resources.DistributionNotFound:
            print(f"‚ùå {pacchetto:<20} NON TROVATO (min: {versione_min})")
        except Exception as e:
            print(f"‚ùå {pacchetto:<20} ERRORE: {e}")
    
    print("=" * 55)

controlla_versioni()
```
